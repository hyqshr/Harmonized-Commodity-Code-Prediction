## keywords from each description -- Main track

```python
def extract_keywords(text, num_keywords=10):
    return ', '.join([kw[0] for kw in keybert_model.extract_keywords(text, keyphrase_ngram_range=(1, 2), top_n=num_keywords)])

# Extract keywords for each description and create a pandas Series
keywords_series = pd.Series([extract_keywords(desc) for desc in df['Description']])
```

```python
df['Keywords'] = keywords_series
print(df)
```

```python
df['HS_Code'].value_counts()
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Flatten, Dense

max_words = 50000
tokenizer = Tokenizer(split=',')
```

```python
df['Keywords'] = df['Keywords'].astype(str)
```

```python
import numpy as np
```

```python
tokenizer.fit_on_texts(df['Keywords'])
sequences = tokenizer.texts_to_sequences(df['Keywords'])

max_length = max([len(seq) for seq in sequences])
vocab_size = len(tokenizer.word_index) + 1

X = pad_sequences(sequences, maxlen=max_length)
y = np.array(df['HS_Code'])

embedding_dim = 50

model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_length))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(509, activation='softmax'))

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train_encoded, y_test_encoded = train_test_split(X, y_encoded, test_size=0.3, random_state=42, stratify=y_encoded)
X_test, X_val, y_test_encoded, y_val_encoded = train_test_split(X_test, y_test_encoded, test_size=0.5, random_state=42)
```

```python
print(X_train[:5])
print(y_train_encoded[:5])
```

```python
model.fit(X_train, y_train_encoded, epochs=10, validation_data=(X_val, y_val_encoded))
```

```python
y_pred_probs = model.predict(X_test)
```

```python
y_pred = np.argmax(y_pred_probs, axis=1)
```

```python
loss, accuracy = model.evaluate(X_test, y_test_encoded)
print(f"Test Loss: {loss}")
print(f"Test Accuracy: {accuracy}")
```

```python
from sklearn.metrics import accuracy_score
print("Test Accuracy:", accuracy_score(y_test_encoded, y_pred))
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

precision = precision_score(y_test_encoded, y_pred, average='weighted')
recall = recall_score(y_test_encoded, y_pred, average='weighted')
f1 = f1_score(y_test_encoded, y_pred, average='weighted')

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
```

```python
from sklearn.metrics import classification_report

report = classification_report(y_test_encoded, y_pred)
print(report)
```
